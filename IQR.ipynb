```python
import pandas as pd
import numpy as np
```

```python
obj = {
    'student_id':['s1','s2','s3','s4','s5','s6','s7'],
    'stat_score':[95,95,85,93,46,95,10],
    'math_score':[65,75,55,23,86,55,10],
}

df = pd.DataFrame(obj)
```

```python
df
```

```python
q1 = df['stat_score'].quantile(0.25) # 제 1사분위수
q3 = df['stat_score'].quantile(0.75) # 제 3사분위수
```

```python
iqr = q3 - q1 # IQR

## 최소 = q1 - 1.5 * iqr / 최대 = q3 + 1.5 * iqr
lower = q1 - 1.5 * iqr
upper = q3 + 1.5 * iqr
```

```python
df[(df['stat_score'] < lower) | (df['stat_score'] > upper)]
```
