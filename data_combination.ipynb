```python
import pandas as pd
import numpy as np
```

```python
obj1 = {
    'id' : ['s1','s2','s3','s4','s5','s6'],
    'score' : [50,70,90,80,60,40]
}

df1 = pd.DataFrame(obj1)

df1
```

```python
obj2 = {
    'id' : ['t1','t2','t3','t4','t5','t6'],
    'score' : [50,70,90,80,60,40]
}

df2 = pd.DataFrame(obj2)

df2
```

```python
# 두 데이터 프레임을 행으로 붙임
pd.concat([df1 , df2] , axis = 0)
```

```python
# 두 데이터 프레임은 열로 붙임
pd.concat([df1,df2] , axis = 1)
```

```python
## pd.merge(df , how = 병합할 유형 ,on = 컬럼명)

pd.merge(left = df1 , right = df2 , how = 'inner' , on = 'score')
```

```python
## groupby(by , as_index = True/False)

obj = {
    'student_id' : ['s1','s2','s3','s4','s5','s6'],
    'stat_score' : [55,90,85,71,63,99],
    'math_score' : [65,99,87,75,57,88],
    'gender' : ['F','M','F','F','M','M'],
    'pre_level': ['B','A','B','B','C','A']
}


df = pd.DataFrame(obj)
```

```python
df
```

```python
df.groupby(by = 'gender' , as_index = False)['stat_score'].sum()
```

```python
df.groupby(by = 'gender' , as_index = False)[['stat_score' , 'math_score']].mean()
```

```python
## group by + agg 를 통해 여러가지 집계 함수를 구할수도 있다.

df.groupby(by = 'pre_level' , as_index = False)[['stat_score' , 'math_score']].agg(['mean' , 'sum'])
```

```python
## map 을 사용한 라벨 인코딩 가능

df['pre_level_map'] = df['pre_level'].map({'A':0 , 'B':1 , 'C':2})
```

```python
df
```

```python
## apply : 사용자 지정 함수 나 집계함수 를 적용 시킬수 있음

df[['stat_score','math_score']].apply(np.sum)
```
