```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import numpy as np
```

```python
obj = {'Class' : ['A','B','C','D','A','C','G']}

df = pd.DataFrame(obj)
df
```

```python
encoder = LabelEncoder()
df['Class'] = encoder.fit_transform(df['Class'])
df

print(encoder.inverse_transform(df['Class']))
```

```python
# numpy select = Label encoding
conditionlist = [
    (df['Class'] == 'A'),
    (df['Class'] == 'B'),
    (df['Class'] == 'C'),
    (df['Class'] == 'D'),
    (df['Class'] == 'G'),
]

choicelist = list(range(5))

np.select(condlist=conditionlist , choicelist = choicelist)
```

```python
## 실습
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder , OneHotEncoder

obj = {
    'student_id' :['s1','s2','s3','s4','s5','s6'],
    'score' : [55,90,85,71,63,99],
}

df = pd.DataFrame(obj)
```

```python
df['grade'] = pd.cut(df['score'],
                     bins = [0,60,70,80,90,100],# n+1
                     labels = ['가','양','미','우','수'])
```

```python
df['grade']
```

```python
lb_encoder = LabelEncoder()
lb_encoder.fit(df['grade'])
lb_labels = lb_encoder.transform(df['grade'])

print(lb_labels)
```

```python
print(lb_encoder.classes_)
```

```python
lb_labels_2d = lb_labels.reshape(-1,1)
```

```python
oh_encoder = OneHotEncoder(sparse = False)
```

```python
oh_encoder.fit(lb_labels_2d)
```

```python
oh_labels = oh_encoder.transform(lb_labels_2d)
```
