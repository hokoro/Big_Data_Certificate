```python
import pandas as pd
from sklearn.datasets import load_diabetes
```

```python
diabets = load_diabetes()
```

```python
df = pd.DataFrame(diabets.data , columns = diabets.feature_names)
```

```python
df.head()
```

```python
from scipy import stats
```

```python
# stats.pearsonr(x = df1 , y = df2)
stats.pearsonr(x = df['age'] , y = df['bmi'])
```

```python
stats.pearsonr(x = df['sex'] , y = df['bmi'])
```

```python
# corr 함수로 사용해도 된다. -> 단순 상관계수 산출은
df.corr()
```

```python
df[['age' , 'bmi']].corr()
```

```python
# 스피어만 상관계수는 순위 사용 , 이상치에 되게 민감하지 않음  , 변수가 서열 척도일떄 사용
stats.spearmanr(df['age'] , df['bmi'])
```

```python
df[['age' , 'bmi']].corr(method = 'spearman')
```
