```python
from sklearn.datasets import load_diabetes
```

```python
import pandas as pd
```

```python
from scipy.stats import linregress

diabetes = load_diabetes()
```

```python
df = pd.DataFrame(diabetes.data , columns = diabetes.feature_names)
```

```python
model = linregress(x = df['bmi'] , y = diabetes.target)
```

```python
model
```

```python
## slope = 독립변수에 대한 추청된 회귀 계수
## intercept = 상수항
## rvalue = 모형의 설명력
## pvalue = 기울기에 대한 통계적 유의성
## stderr , intercept_stderr = 회귀계수에 대한 표준오차를 의미한다
```

```python
model.slope
```

```python
## 선형 회귀
from sklearn.linear_model import LinearRegression

columns = ['bmi','bp','s1','s2','s3']
x = df[columns]
y = diabetes.target
```

```python
model = LinearRegression()
```

```python
model.fit(x,y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model.score(x,y)
```

```python
## 릿지 회귀
from sklearn.linear_model import Ridge

model = Ridge(alpha = 0.1)
```

```python
model.fit(x,y)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
## 라쏘 회귀
from sklearn.linear_model import Lasso
model = Lasso(alpha = 0.5)

model.fit(x,y)
```

```python
model.coef_
```

```python
model.intercept_
```
