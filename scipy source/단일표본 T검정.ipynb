문제: 다음은 22명의 학생들이 국어시험에서 받은 점수이다. 학생들의 평균이 75보다 크다고 할 수 있는가?
- 귀무가설(H0): 모평균은 mu와 같다. (μ = mu), 학생들의 평균은 75이다
- 대립가설(H1): 모평균은 mu보다 크다. (μ > mu), 학생들의 평균은 75보다 크다
- 가정:모집단은 정규분포를 따른다.
- 표본의 크기가 충분히 크다.

```python
from scipy import stats
```

```python
# 데이터
scores = [75, 80, 68, 72, 77, 82, 81, 79, 70, 74, 76, 78, 81, 73, 81, 78, 75, 72, 74, 79, 78, 79]

# 모평균 가설검정
mu = 75  # 검정할 모평균
alpha = 0.05  # 유의수준
```

```python
# 단일 표본 T 검정은 ttest_1samp

st,pv = stats.ttest_1samp(scores ,mu , alternative = 'greater') # 모평균 보다 크다 ? -> alternative = 'greater'

print(f'검정 통계량 : {st}')
print(f'p-value : {pv}')

```

```python
if pv < 0.05:
    print('모평균은 75점보다 큽니다')
else:
    print('모평균은 75점과 같습니다 ')
```
