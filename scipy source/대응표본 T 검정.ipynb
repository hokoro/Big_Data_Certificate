주어진 데이터는 고혈압 환자 치료 전후의 혈압이다. 해당 치료가 효과가 있는지 대응(쌍체)표본 t-검정을 진행하시오

- 귀무가설(H0):  μ>= 0
- 대립가설(H1):  μ < 0
- μ = (치료 후 혈압 - 치료 전 혈압)의 평균
- 유의수준: 0.05

- μ의 표본평균은?(소수 둘째자리까지 반올림)
- 검정통계량 값은?(소수 넷째자리까지 반올림)
- p-값은?(소수 넷째자리까지 반올림)
- 가설검정의 결과는? (유의수준 5%)

```python
import pandas as pd

df = pd.read_csv("/content/drive/MyDrive/빅데이터 분석기사/data/high_blood_pressure.csv")
```

```python
df.head()
```

```python
df['diff'] = df['bp_post'] - df['bp_pre']
```

```python
# 소수점 둘째 자리 까지 반올림
print(round(df['diff'].mean(),2))
```

## 대응표본 T 검정

- alternative = less (적음)
- alternative = two-sided (양면)
- alternative = greater(더 큼 )

```python
from scipy import stats

st , pv = stats.ttest_rel(df['bp_post'] , df['bp_pre'] , alternative='less')
print(round(st,4)) # 검정 통계량
```

```python
print(round(pv,4))

if pv < 0.05:
    print(f'{round(pv,4)} 이기 떄문에 대립가설 채택')
else:
    print('귀무가설 채택')
```
