## 독립표본 T 검정
- 어떤 특정 약물을 복용한 사람들의 평균 체온이 복용하지 않은 사람들의 평균 체온과 유의미하게 다른지 검정해보려고 합니다.
- 가정:약물을 복용한 그룹과 복용하지 않은 그룹의 체온 데이터가 각각 주어져 있다고 가정합니다.
- 각 그룹의 체온은 정규분포를 따른다고 가정합니다.

```python
from scipy import stats
```

```python
# 대응과 독립의 차이
# 대응 : ex) 한사람의 변화  독립 : 두 집단의 독립적인 변화
group1 = [36.8, 36.7, 37.1, 36.9, 37.2, 36.8, 36.9, 37.1, 36.7, 37.1] # 약물 복용 O
group2 = [36.5, 36.6, 36.3, 36.6, 36.9, 36.7, 36.7, 36.8, 36.5, 36.7] # 약물 복용 X
```

```python
st , pv = stats.ttest_ind(group1 , group2)

print(f'검정 통계량 :{st}')
print(f'p-value: {pv}')
```

```python
if round(pv ,4) < 0.05:
    print('약물을 복용한 사람과 복용하지 않은 사람들의 유의미한 차이가 있다.')
else:
    print('약물을 복용한 사람과 복용하지 않은 사람들의 유의미한 차이가 없다')
```
