```python
from sklearn.tree import DecisionTreeClassifier , DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_breast_cancer , load_iris , load_diabetes
from sklearn.metrics import roc_curve , auc , f1_score , mean_squared_error
```

```python
breast = load_breast_cancer()
data = breast.data
target = breast.target

train_x , test_x , train_y , test_y = train_test_split(data , target ,test_size = 0.2 , random_state = 42 , stratify = target)
```

```python
model = DecisionTreeClassifier(max_depth = 3 , min_samples_leaf = 10 , random_state = 42)
```

```python
model.fit(train_x , train_y)
```

```python
y_score = model.predict_proba(test_x)[:,1]
```

```python
fpr,tpr,thresholds = roc_curve(test_y , y_score)
```

```python
AUC = auc(fpr , tpr)
```

```python
AUC
```

```python
iris = load_iris()

data = iris.data
target = iris.target
```

```python

train_x , test_x , train_y , test_y = train_test_split(data , target ,test_size = 0.2 , random_state = 42 , stratify = target)
```

```python
model = DecisionTreeClassifier(max_depth = 3 , min_samples_leaf = 10 , random_state = 42)

model.fit(train_x , train_y)
```

```python
y_pred = model.predict(test_x)
```

```python
macro_f1 = f1_score(test_y , y_pred , average = 'macro')
```

```python
macro_f1
```

```python
diabetes = load_diabetes()
```

```python
data = diabetes.data
target = diabetes.target


train_x , test_x , train_y , test_y = train_test_split(data , target ,test_size = 0.2 , random_state = 42)
```

```python
model = DecisionTreeRegressor(max_depth = 3 , min_samples_leaf = 10 , random_state = 42)
model.fit(train_x , train_y)
```

```python
y_pred = model.predict(test_x)
```

```python
print(mean_squared_error(test_y , y_pred , squared = False))
print(mean_squared_error(test_y , y_pred) ** 0.5)
```
