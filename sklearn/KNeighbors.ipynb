```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split

from sklearn.datasets import load_breast_cancer

breast = load_breast_cancer()
data = breast.data
target = breast.target
```

```python
train_x , test_x , train_y , test_y = train_test_split(data , target , test_size = 0.2 , random_state = 42 , stratify = target)
```

```python
model = KNeighborsClassifier(n_neighbors = 5 , weights = 'distance') #weight = uniform(균일 가중치) / distance(거리 역수 )
```

```python
model.fit(train_x , train_y)
```

```python
from sklearn.metrics import roc_curve , auc

y_score = model.predict_proba(test_x)[:,1]
```

```python
fpr,tpr,thresholds = roc_curve(test_y , y_score)
```

```python
AUC = auc(fpr ,tpr)
```

```python
from sklearn.datasets import load_iris

iris = load_iris()

data = iris.data
target = iris.target
```

```python
train_x , test_x , train_y , test_y = train_test_split(data , target , test_size = 0.2 , random_state = 42 , stratify = target)
```

```python
model = KNeighborsClassifier(n_neighbors = 10 , weights = 'distance')

model.fit(train_x , train_y)
```

```python
from sklearn.metrics import f1_score
y_pred = model.predict(test_x)

macro_f1 = f1_score(test_y , y_pred , average = 'macro')

print(macro_f1)
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import train_test_split

from sklearn.datasets import load_diabetes
```

```python
diabetes = load_diabetes()

data = diabetes.data
target = diabetes.target

train_x , test_x , train_y ,test_y = train_test_split(data , target , test_size = 0.2 , random_state = 42)
```

```python
model = KNeighborsRegressor(n_neighbors = 5 , weights = 'distance')
```

```python
model.fit(train_x,train_y)
```

```python
from sklearn.metrics import mean_squared_error

y_pred = model.predict(test_x)

print(mean_squared_error(test_y , y_pred) ** 0.5)
print(mean_squared_error(test_y , y_pred , squared = False))
```
