```python
## MLP Classifier
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split

from sklearn.datasets import load_breast_cancer , load_iris , load_diabetes
```

```python
breast = load_breast_cancer()

data = breast.data
target = breast.target

train_x , test_x , train_y , test_y = train_test_split(data , target , test_size = 0.2 , random_state = 42 , stratify = target)
```

```python
model = MLPClassifier(alpha = 0.5 , max_iter = 500 , random_state = 42)
```

```python
model.fit(train_x , train_y)
```

```python
from sklearn.metrics import roc_curve , auc
```

```python
y_score = model.predict_proba(test_x)[:,1]
```

```python
fpr,tpr,thresholds = roc_curve(test_y , y_score)
```

```python
AUC = auc(fpr,tpr)
```

```python
AUC
```

```python
iris = load_iris()
data = iris.data
target = iris.target
```

```python
train_x , test_x , train_y , test_y = train_test_split(data , target , test_size = 0.2 , random_state = 42 , stratify = target)
```

```python
model = MLPClassifier(alpha = 0.5 , max_iter = 500 , random_state = 42)
```

```python
model.fit(train_x , train_y)
```

```python
y_pred = model.predict(test_x)
```

```python
from sklearn.metrics import f1_score

macro_f1 = f1_score(test_y, y_pred , average = 'macro')
```

```python
macro_f1
```

```python
## MLP Regressor
from sklearn.neural_network import MLPRegressor
```

```python
diabetes = load_diabetes()

data = diabetes.data
target = diabetes.target

train_x , test_x , train_y , test_y = train_test_split(data , target , test_size = 0.2 , random_state = 42)
```

```python
model = MLPRegressor(alpha = 0.5 , max_iter = 10000 , random_state = 42)
```

```python
model.fit(train_x , train_y)
```

```python
from sklearn.metrics import mean_squared_error

y_pred = model.predict(test_x)

print(mean_squared_error(test_y , y_pred , squared = False))
print(mean_squared_error(test_y , y_pred) ** 0.5)
```
