```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
df = pd.read_csv('/content/drive/MyDrive/빅데이터 분석기사/data/USArrests.csv' , index_col = 0)
```

```python
df.head()
```

```python
x = StandardScaler().fit_transform(df)
```

```python
# 주성분 분석은 전체 컬럼수를 기준으로 유의미한 변수를 요약하고 축소하는 기법
pca = PCA(n_components = 4)
```

```python
pca_arr = pca.fit_transform(x)
```

```python
pca_arr
```

```python
result = {
    '분산 기여율' : pca.explained_variance_ratio_,
    '누적 기여율' : pca.explained_variance_ratio_.cumsum()
}
```

```python
result = pd.DataFrame(result , index = ['PCA1', 'PCA2', 'PCA3' , 'PCA4']).T
```

```python
print(result)
```

```python
pca = PCA(n_components = 2)
pca_arr = pca.fit_transform(x)
df_pca = pd.DataFrame(pca_arr , columns = ['PCA1' , 'PCA2'])
df_pca.head()
```
