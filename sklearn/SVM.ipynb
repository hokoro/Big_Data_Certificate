```python
## SVC
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
```

```python
from sklearn.datasets import load_breast_cancer
breast_cancer = load_breast_cancer()
data = breast_cancer.data
target = breast_cancer.target
```

```python
train_x , test_x , train_y ,test_y = train_test_split(data , target , test_size =0.2 ,random_state = 42 , stratify = target)
```

```python
model = SVC(kernel = 'linear', C = 0.5 , probability = True) #kernel 은 커널 함수를 어떤걸 쓸거냐 , C 는 규제 강도 ,probability 는 predict_proba 를 쓸라면 반드시 True 를 해야한다 SVC 에서

model.fit(train_x , train_y)
```

```python
from sklearn.metrics import roc_curve , auc
```

```python
y_score = model.predict_proba(test_x)[:,1]
```

```python
fpr,tpr,thresholds = roc_curve(test_y ,y_score)
```

```python
AUC = auc(fpr , tpr)
```

```python
print(AUC)
```

```python
##linearsvc
from sklearn.svm import LinearSVC
from sklearn.model_selection import train_test_split

from sklearn.datasets import load_iris

iris = load_iris()
data = iris.data
target = iris.target

train_x , test_x , train_y , test_y = train_test_split(data , target , test_size = 0.2 , random_state = 42 , stratify = target)
```

```python
model = LinearSVC(C = 0.1)
```

```python
model.fit(train_x, train_y)
```

```python
from sklearn.metrics import f1_score

y_pred = model.predict(test_x)

macro_f1 = f1_score(test_y , y_pred , average='macro')
```

```python
print(macro_f1)
```

```python
train_x , test_x , train_y ,test_y = train_test_split(data, target , test_size = 0.2 , random_state = 42)
```

```python
model = SVR(C = 0.1 , epsilon = 0.01)
model2 = LinearSVC(C = 0.1 , loss = 'squared_hinge' )
```

```python
from sklearn.metrics import mean_squared_error

y_pred = model.predict(test_x)
y_pred2 = model2.predict(test_x)
rmse = mean_squared_error(test_y , y_pred) ** 0.5
rmse2 = mean_squared_error(test_y , y_pred) ** 0.5
```
