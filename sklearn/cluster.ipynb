- 정답 즉 레이블 타겟이 모를때는 실루엣 계수 (silhouette) / 아는 경우 랜드 지수 (Rand Index) 와 조정 랜드 지수 (Adjust Rand Index)

```python
## sklearn.metrics.rand_score(y_true , y_pred)
## sklearn.adjusted_rand_score(y_true , y_pred)

y_true = [0,0,0,1,1,1,1,2,2]
y_pred = [0,0,1,1,1,1,2,2,2]

from sklearn.metrics.cluster import rand_score , adjusted_rand_score

ri = rand_score(y_true , y_pred)

print(ri)
```

```python
ari = adjusted_rand_score(y_true , y_pred)
print(ari)
```

```python
## 계층적 군집 분석
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import rand_score , adjusted_rand_score
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
data = iris.data
target = iris.target
```

```python
# 와드 연결법
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics.cluster import rand_score , adjusted_rand_score

agg_ward = AgglomerativeClustering(n_clusters = 3,linkage='ward')
ward = agg_ward.fit_predict(data)
```

```python
agg_avg = AgglomerativeClustering(n_clusters = 3 , linkage = 'average')
avg = agg_avg.fit_predict(data)
```

```python
agg_comp = AgglomerativeClustering(n_clusters = 3 , linkage = 'complete')
comp = agg_comp.fit_predict(data)
```

```python
agg_sing = AgglomerativeClustering(n_clusters=3 , linkage = 'single')
sing = agg_sing.fit_predict(data)
```

```python
# RI 비교 = 아는 경우
print(rand_score(target ,  ward))
print(rand_score(target , avg))
print(rand_score(target , comp))
print(rand_score(target , sing))
```

```python
# ARS
print(adjusted_rand_score(target , ward))
print(adjusted_rand_score(target , avg))
print(adjusted_rand_score(target , comp))
print(adjusted_rand_score(target , sing))
```

```python
## kmeans
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score , silhouette_samples

kmeas_k2 = KMeans(n_clusters = 2 , random_state = 2022)
pred_kw2 = kmeas_k2.fit_predict(data)
```

```python
kmeas_k3 = KMeans(n_clusters = 3 , random_state = 2022)
pred_kw3 = kmeas_k3.fit_predict(data)
```

```python
kmeas_k4 = KMeans(n_clusters = 4 , random_state = 2022)
pred_kw4 = kmeas_k4.fit_predict(data)
```

```python
import pandas as pd

sil_k2 = silhouette_samples(data , pred_kw2)
```

```python
sil_k2
```

```python
df_k2 = pd.DataFrame({'labels':pred_kw2 , 'silhouette':sil_k2})
```

```python
df_k2
```

```python
print(df_k2.groupby('labels')['silhouette'].mean())
```

```python
sil_k3 = silhouette_samples(data , pred_kw3)
```

```python
df_k3 = pd.DataFrame({'labels':pred_kw3 , 'silhouette':sil_k3})
```

```python
print(df_k3.groupby('labels')['silhouette'].mean())
```

```python
silhouette_score(data , pred_kw3)
```
