```python
from sklearn.datasets import load_breast_cancer
breast_cancer = load_breast_cancer()
```

```python
x = breast_cancer.data
y = breast_cancer.target
```

```python
from sklearn.model_selection import train_test_split

'''
test_size = train/test 를 나누는 test의 비율
random_state = 랜덤 분할을 할떄 seed 값
x,y = x데이터 , y데이터
'''
train_x,test_x,train_y,test_y = train_test_split(x,y,test_size = 0.2,random_state = 42)
```

```python
train_x.shape , train_y.shape , test_x.shape,test_y.shape
```

```python
## stratify = target(컬럼의 0과 1의 비율을 반영하여 데이터 분할)
y
```

```python
import numpy as np
zero = 0
one = 0
for i in y:
    if i == 0:
        zero+= 1
    else:
        one+=1

print(zero , one)
```

```python
## 위와 같이 0과 1의 비율을 맞춰주기 위해서 train_test_split 에 stratify 옵션을 사용한다
train_x,test_x,train_y,test_y = train_test_split(x,y,test_size = 0.2 , random_state = 42 , stratify = y)
```

```python
from sklearn.model_selection import KFold
```

```python
## KFold 는 전체데이터의 test_data와 train_data 를 반복하면서 학습하는 방법이다.
import numpy as np
```

```python
x = np.arange(10)
```

```python
kfold = KFold(n_splits = 5) # 5개로 분할한다는것은 10 / n_split 이 각 분할마다 데이터수이다 즉 10/5 = 2
for train_idx , test_idx in kfold.split(x):
    print(f'학습 : {train_idx} | 평가 : {test_idx}')
```

```python
# 분할 하기전에 shuffle 하고 싶으면 shuffle 옵션을 사용한다.
kfold2 = KFold(n_splits=5 , shuffle= True)
for train_idx , test_idx in kfold.split(x):
    print(f'학습 : {train_idx} | 평가 : {test_idx}')
```

```python
import numpy as np
x = np.arange(15)
y = [0] * 6 + [1] * 3 + [2] * 6

# 각 y별 데이터 를 split 하여 계산한다 0 * 2  + 1 * 1 + 2 * 2
from sklearn.model_selection import StratifiedKFold
kfold = StratifiedKFold(n_splits = 3)
```

```python

for train_idx , test_idx in kfold.split(x,y):
    print(f'학습 : {train_idx} | 평가 : {test_idx}')
```
