## 문제 1
- iris 데이터 셋을 불러와 Sepal.Width 칼럼에 대해 Sepal.Width 의 평균값을 기준으로 3배 표준편차 이상으로 떨어진 값들의 합을 구하여라

```python
from sklearn.datasets import load_iris
import numpy as np
import pandas as pd


iris = load_iris()

iris
```

```python
df = pd.DataFrame(data = iris['data'] , columns = iris['feature_names'])
```

```python
df
```

```python
df.isnull().sum()
```

```python
mean_ = df['sepal width (cm)'].mean()
```

```python
std_ = df['sepal width (cm)'].std()
```

```python
mean_ , std_
```

```python
low = mean_ - (3 * std_)
high = mean_ + (3 * std_)
```

```python
df.loc[(df['sepal width (cm)']<low) | (df['sepal width (cm)'] >high) , 'sepal width (cm)'].sum()
```
