## iris에서 Sepal Length와 Sepal Width의 상관계수 계산하고 소수 둘째자리까지 출력하시오

```python
import pandas as pd
from sklearn.datasets import load_iris
```

```python
# iris 데이터셋 로드
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
# 상관계수를 구할떄는 df.corr() 를 변수로 저장 -> loc 를 통해 위치 지정을 통해 컬럼 두개를 선언

corr = df.corr()

print(round(corr.loc['sepal length (cm)', 'sepal width (cm)'],2))
```
