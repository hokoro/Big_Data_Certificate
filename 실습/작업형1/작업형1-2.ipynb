## 문제2
- 앞에서 순서대로 70% 데이터만 활용하여
- 'f1' 컬럼 결측피를 중앙값으로 채우기 전후의 표준편차를 구하고
- 두 표준편차 차이 계산하기

```python
df = pd.read_csv('/content/drive/MyDrive/빅데이터 분석기사/data/basic1.csv')
```

```python
df_len = df.shape[0]
df_len
```

```python
df_len = int(df_len * 0.7)
df_len
```

```python
df_2 = df.iloc[:df_len,:]
```

```python
df_2
```

```python
before_f1_std = df_2['f1'].std()
before_f1_std
```

```python
after_f1_std = df_2['f1'].fillna(df_2['f1'].median()).std()
```

```python

after_f1_std
```

```python
print(before_f1_std - after_f1_std)
```

```python
df.info()
```

```python
age = df['age']
```

```python
Q1 = age.quantile(0.25)
Q3 = age.quantile(0.75)

range_ = age.std() * 1.5
```

```python
range_
```

```python
mean = age.mean()
```
