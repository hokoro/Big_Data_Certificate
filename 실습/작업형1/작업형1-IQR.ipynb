```python
Q3 = train['Fare'].quantile(0.75)
Q1 = train['Fare'].quantile(0.25)

IQR = Q3 - Q1
```

```python
IQR
```

```python
fare = train['Fare']
```

```python
low = Q1['Fare'] - 1.5*IQR
high = Q3['Fare'] + 1.5 * IQR
```
