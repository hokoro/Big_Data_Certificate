## 결측치 처리
- 주어진 데이터에서 결측치가 80%이상 되는 컬럼은(변수는) 삭제하고, 80% 미만인 결측치가 있는 컬럼은 'city'별 중앙값으로 값을 대체하고 'f1'컬럼의 평균값을 출력하세요!

```python
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/빅데이터 분석기사/data/Nulldata/basic1.csv')
```

```python
df
```

```python
na_count = df.isnull().sum()
```

```python
all_nums = df.shape[0]
all_nums
```

```python
cols = df.columns
```

```python
drop_columns = []
for col in cols:
    print(f'컬럼 이름 : {col}/{na_count[col]}')
    print(f'결측치 비율 : {(na_count[col] / all_nums)}')
    drop_rate = (na_count[col] / all_nums)
    if drop_rate >= 0.8:
        drop_columns.append(col)
```

```python
drop_columns
```

```python
df = df.drop(drop_columns,axis = 1)
```

```python
df
```

```python
f1_city = df.groupby('city')['f1'].median()
f1_city
```

```python
df.loc[(df['city'] == '서울') & (df['f1'].isnull()) , 'f1'] = f1_city['서울']
df.loc[(df['city'] == '경기') & (df['f1'].isnull()) , 'f1'] = f1_city['경기']
df.loc[(df['city'] == '대구') & (df['f1'].isnull()) , 'f1'] = f1_city['대구']
df.loc[(df['city'] == '부산') & (df['f1'].isnull()) , 'f1'] = f1_city['부산']
```
