## T1-30. 12월 25일 결제 금액(price)은 12월 총 결제금액의 몇 %인가? (정수로 출력)

```python
import pandas as pd
import numpy as np

df = pd.read_csv('/content/drive/MyDrive/빅데이터 분석기사/data/payment/payment.csv')
```

```python
df.info()
```

```python
df['date'] = pd.to_datetime(df['date'] , format = '%Y%m%d')
```

```python
df
```

```python
price_12 = df.loc[(df['date'].dt.month == 12) &(df['date'].dt.day == 25) , 'price'].sum()
```

```python
price_all = df.loc[(df['date'].dt.month == 12) , 'price'].sum()
```

```python
print(int((price_12 / price_all) * 100))
```
