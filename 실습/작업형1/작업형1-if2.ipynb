## 문제
- city와 f4를 기준으로 f5의 평균값을 구한 다음, f5를 기준으로 상위 7개 값을 모두 더해 출력하시오 (소수점 둘째자리까지 출력)
- 데이터 : basic1.csv

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/drive/MyDrive/빅데이터 분석기사/data/IQR2/basic1.csv')
```

```python
df
```

```python
df.isnull().sum()
```

```python
df['f4']
```

```python
df['f5']
```

```python
f4_f5_mean = df.groupby(['city','f4'] , as_index = False)[['f5']].mean()
```

```python
f4_f5_mean = f4_f5_mean.sort_values('f5',ascending = False)
```

```python
f4_f5_mean
```

```python
answer = f4_f5_mean.iloc[:7]
answer
```

```python
print(np.round(sum(answer['f5']),2))
```
