## min-max스케일링 기준 상하위 5% 구하기
- 주어진 데이터에서 'f5'컬럼을 min-max 스케일 변환한 후, 상위 5%와 하위 5% 값의 합을 구하시오

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
import pandas as pd
import numpy as np

df = pd.read_csv('/content/drive/MyDrive/빅데이터 분석기사/data/min_max_scale/basic1.csv')
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import MinMaxScaler

mm = MinMaxScaler()

df['f5'] = mm.fit_transform(df[['f5']])
```

```python
low_5 = df['f5'].quantile(0.05)
high_5 = df['f5'].quantile(0.95)
```

```python
print(low_5 + high_5)
```

```python
df2 = df.copy()
```

```python
min_f5 = df2['f5'].min()
max_f5 = df2['f5'].max()
```

```python
df2['f5'] = (df2['f5'] - min_f5) / (max_f5 - min_f5)
```

```python
low_5 = df2['f5'].quantile(0.05)
high_5 = df2['f5'].quantile(0.95)
```

```python
print(low_5 + high_5)
```
