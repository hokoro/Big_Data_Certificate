## 왜도와 첨도 구하기
- 주어진 데이터 중 train.csv에서 'SalePrice'컬럼의 왜도와 첨도를 구한 값과, 'SalePrice'컬럼을 스케일링(log1p)로 변환한 이후 왜도와 첨도를 구해 모두 더한 다음 소수점 2째자리까지 출력하시오

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/drive/MyDrive/빅데이터 분석기사/data/skew and kurtosis/train.csv')
df
```

```python

salesprice_skew = df['SalePrice'].skew()
salesprice_kurtosis = df['SalePrice'].kurtosis()
```

```python
salesprice_skew , salesprice_kurtosis
```

```python
df['SalePrice'] = np.log1p(df['SalePrice'])
```

```python
df['SalePrice']
```

```python
salesprice_skew_log = df['SalePrice'].skew()
salesprice_kurtosis_log = df['SalePrice'].kurtosis()
```

```python
salesprice_skew_log , salesprice_kurtosis_log
```

```python
print(np.round(salesprice_skew + salesprice_kurtosis + salesprice_skew_log + salesprice_kurtosis_log,2))
```
