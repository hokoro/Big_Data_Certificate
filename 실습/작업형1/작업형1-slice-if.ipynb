## 문제
- 주어진 데이터 셋에서 age컬럼 상위 20개의 데이터를 구한 다음
- f1의 결측치를 중앙값으로 채운다.
- 그리고 f4가 ISFJ와 f5가 20 이상인
- f1의 평균값을 출력하시오!
- 데이터셋 : basic1.csv

```python
import pandas as pd
import numpy as np

df = pd.read_csv('/content/drive/MyDrive/빅데이터 분석기사/data/basic1/basic1.csv')
```

```python
df
```

```python
df.isnull().sum()
```

```python
df = df.sort_values('age' , ascending = False)
```

```python
df_age_20 = df[:20]
```

```python
df_age_20
```

```python
df_age_20['f1'] = df_age_20['f1'].fillna(df_age_20['f1'].median())
```

```python
df_age_20
```

```python
print(df_age_20.loc[(df_age_20['f4'] == 'ISFJ') & (df_age_20['f5'] >=20) , 'f1'].mean())
```
