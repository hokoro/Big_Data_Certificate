## 조건에 맞는 값 바꾸고 출력
- 'f4'컬럼의 값이 'ESFJ'인 데이터를 'ISFJ'로 대체하고, 'city'가 '경기'이면서 'f4'가 'ISFJ'인 데이터 중 'age'컬럼의 최대값을 출력하시오!

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/drive/MyDrive/빅데이터 분석기사/data/value change/basic1.csv')
```

```python
df.head()
```

```python
df.loc[df['f4'] == 'ESFJ' , 'f4'] = 'ISFJ'
```

```python
print(df.loc[(df['f4'] == 'ISFJ') & (df['city'] == '경기') , 'age'].max())
```
