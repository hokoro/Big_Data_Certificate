## 문제
- 주어진 데이터 셋에서 f2가 0값인 데이터를 age를 기준으로 오름차순 정렬하고
- 앞에서 부터 20개의 데이터를 추출한 후
- f1 결측치(최소값)를 채우기 전과 후의 분산 차이를 계산하시오 (소수점 둘째 자리까지)

- 데이터셋 : basic1.csv

```python
df = pd.read_csv('/content/drive/MyDrive/빅데이터 분석기사/data/basic1/basic1.csv')
df
```

```python
df_f2_zero = df.loc[df['f2'] == 0 , :]
```

```python
df_f2_zero = df_f2_zero.sort_values('age',ascending = True)
df_f2_zero
```

```python
df_f2_zero = df_f2_zero.iloc[:20]
df_f2_zero
```

```python
df_f2_zero.isnull().sum()
```

```python
before = df_f2_zero['f1'].var()
```

```python
df_f2_zero['f1'] = df_f2_zero['f1'].fillna(df_f2_zero['f1'].min())
```

```python
np.round((before - after) , 2)
```

```python
cond = (df['f2']==0)
df = df[cond].sort_values('age', ascending=True).reset_index(drop=True)

# 앞에서 부터 20개의 데이터
df = df[:20]
print(df)
# f1 결측치(최소값)를 채우기 전과 후의 분산
df_var1 = df['f1'].var()
df['f1'] = df['f1'].fillna(df['f1'].min())
df_var2 = df['f1'].var()

print(df_var1 , df_var2)
# 소수점 둘째자리까지 출력
print(round(df_var1 - df_var2, 2))
```
