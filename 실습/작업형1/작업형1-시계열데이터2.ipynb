- 주어진 데이터에서 2022년 5월 주말과 평일의 sales컬럼 평균값 차이를 구하시오 (소수점 둘째자리까지 출력, 반올림)

```python
import pandas as pd
import numpy as np
```

```python
a1 = df.loc[(df['Date'].dt.day_name() == 'Sunday') | (df['Date'].dt.day_name() == 'Saturday'), :]
```
