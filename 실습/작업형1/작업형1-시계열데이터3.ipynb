## 문제
- 주어진 데이터에서 2022년 월별 Sales 합계 중 가장 큰 금액과
- 2023년 월별 Sales 합계 중 가장 큰 금액의 차이를 절대값으로 구하시오.
- 단 Events컬럼이 '1'인경우 80%의 Salse값만 반영함
- (최종값은 소수점 반올림 후 정수 출력)
- 데이터셋 : basic2.csv

```python
import numpy as np
```

```python
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/빅데이터 분석기사/data/basic2/basic2.csv')
```

```python
df
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df_2022 = df.loc[df['Date'].dt.year == 2022 , :]
df_2023 = df.loc[df['Date'].dt.year == 2023 , :]
```

```python
df_2022['month'] = df_2022['Date'].dt.month
```

```python
df_2022
```

```python
df_2023['month'] = df_2023['Date'].dt.month
```

```python
df_2022.loc[df_2022['Events'] == 1 , 'Sales'] = df_2022['Sales'] * 0.8
df_2023.loc[df_2023['Events'] == 1 , 'Sales'] = df_2023['Sales'] * 0.8
```

```python
max_2022 = df_2022.groupby('month' , as_index = False)['Sales'].sum()
max_2023 = df_2023.groupby('month' , as_index = False)['Sales'].sum()
```

```python
max_2022 = max_2022['Sales'].max()
```

```python
max_2023 = max_2023['Sales'].max()
```

```python
int(np.round(abs(max_2023 - max_2022)))
```
