## 박스 콕스 변환 (Box-Cox 변환)

## 여존슨 변환

```python
# f1 컬럼 결측치 채우기
df['f1'].fillna(mode_num , inplace = True)
```

```python
f1_list = df['f1'].to_list()
f1_list = np.reshape(f1_list , newshape = (-1,1))
```

```python

yeo_johnson = power_transform(f1_list , method = 'yeo-johnson' ,standardize=False)
box_cox = power_transform(f1_list , method = 'box-cox',standardize=False)
```
