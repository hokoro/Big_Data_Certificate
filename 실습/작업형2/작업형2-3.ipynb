### 결측치 처리

```python
# 결측치 처리
df_train['AnnualIncome'] = df_train['AnnualIncome'].fillna(df_train['AnnualIncome'].mean())
df_test['AnnualIncome'] = df_test['AnnualIncome'].fillna(df_test['AnnualIncome'].mean())
```

## LabelEncoding

```python
df_train.info()
```

```python
cols = df_train.select_dtypes(include = 'object').columns.tolist()
```

```python
cols
```

```python
y = df_train.pop('TravelInsurance')
```

```python
df = pd.concat([df_train , df_test])
```

```python
df
```

```python
df.isnull().sum()
```

```python
cols = df.select_dtypes(include = 'object').columns.tolist()
```

```python
from sklearn.preprocessing import LabelEncoder
for col in cols:
    print(f'object : {col}')
    mm = LabelEncoder()
    df[col] = mm.fit_transform(df[col])
```

```python
from sklearn.preprocessing import LabelEncoder
for col in cols:
    print(f'object : {col}')
    mm = LabelEncoder()
    df_train[col] = mm.fit_transform(df_train[col])
```

```python
df.info()
```

```python
from sklearn.model_selection import train_test_split
train_x , val_x , train_y , val_y = train_test_split(df_train , y , test_size = 0.2 , random_state = 42 , stratify= y)
```

```python
train_x.shape , train_y.shape , val_x.shape ,val_y.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import roc_auc_score
```

```python
model = DecisionTreeClassifier(max_depth = 5 , min_samples_leaf= 10,random_state = 42)
```

```python
model.fit(train_x , train_y)
```

```python
y_pred = model.predict_proba(val_x)
```

```python
print(roc_auc_score(val_y , y_pred[:,1]))
```

```python
y_pred_test = model.predict_proba(df)[:,1]
```

```python
answer = pd.DataFrame({'id':range(df.shape[0]) , 'TravelInsurance':y_pred_test})
```
