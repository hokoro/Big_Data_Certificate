```python
import pandas as pd
train = pd.read_csv('/content/drive/MyDrive/빅데이터 분석기사/data/car2 Regression/train.csv')
test = pd.read_csv('/content/drive/MyDrive/빅데이터 분석기사/data/car2 Regression/test.csv')
```

```python
train
```

```python
y = train['price']
x = train.drop(['price'] , axis = 1)
```

```python
test
```

```python
x
```

```python
y
```

```python
from sklearn.preprocessing import LabelEncoder

cols = x.select_dtypes('object').columns.to_list()
```

```python
for col in cols:
    le = LabelEncoder()
    x[col] = le.fit_transform(x[col])
    test[col] = le.transform(test[col])

```

```python
x
```

```python
test
```

```python
from sklearn.model_selection import train_test_split

train_x,val_x , train_y , val_y = train_test_split(x,y , test_size = 0.2 , random_state = 42)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model = RandomForestRegressor(max_depth = 7 , n_estimators = 100 , min_samples_leaf = 3 , random_state =42)
```

```python
model.fit(train_x , train_y)
```

```python
y_pred = model.predict(val_x)
```

```python
from sklearn.metrics import mean_squared_error

print(mean_squared_error(val_y , y_pred , squared = False))
print(mean_squared_error(val_y,y_pred) ** 0.5)
```

```python
y_pred = model.predict(test)
```

```python
answer = pd.DataFrame({'id':range(1617),'price':y_pred})
```
